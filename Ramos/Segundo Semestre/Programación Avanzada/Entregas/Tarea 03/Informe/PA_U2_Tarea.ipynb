## Ejercicio 1

Realice una clase llamada Circulo, la cual se pueda inicializar con un argumento (el radio), y tenga dos atributos: perimetro y rea.
La idea es que el usuario pueda luego inicializar:  
  
```
     C = Circulo(3)  
     print(C.perimetro, C.area)
```
y arroje los resultados deseados.

````python
import math
````

````python
class Circulo:
    def __init__(self, radio):
        self.radio = radio
        self.perimetro = 2 * math.pi * radio
        self.area = math.pi * radio ** 2
````

````python
# Ejemplo de uso
C = Circulo(3)
print(C.perimetro, C.area)
````

## Ejercicio 2

Escriba la clase Punto3D que permita ingresar un punto de la forma (x,y,z) y se pueda acceder a sus coordenadas como atributo, adems, cree un mtodo (funcin de la clase) llamado distancia que permita pasarle como argumento otro objeto de tipo Punto3D y calcule la distancia Euclidiana entre ellos.

Debe ser capaz de ejecutar estos commandos:

```
     A = Punto3D(1,2,0)
     B = Punto3D(3,0,0)
     print(A.x, A.y, A.distancia(B))
```



````python
import math
````

````python
class Punto3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def distancia(self, otro_punto):
        return math.sqrt((otro_punto.x - self.x) ** 2 +
                         (otro_punto.y - self.y) ** 2 +
                         (otro_punto.z - self.z) ** 2)
````

````python
# Ejemplo de uso
A = Punto3D(1, 2, 0)
B = Punto3D(3, 0, 0)
print(A.x, A.y, A.distancia(B))
````

## Ejercicio 3

Grafique en una misma figura, la funcion $f(x) = x^2 + 2x + 3$ de color rojo con lnea segmentada y $g(x) = x^2 + 1$ con color verde y lnea continua. Debe incluir una leyenda con las funciones que se estn graficando, con ttulo de grfica y con una cuadrcula. Los ejes deben estar etiquetados como Eje x y Eje y, respectivamente.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Definir las funciones
def f(x):
    return x**2 + 2*x + 3

def g(x):
    return -x**2 + 1
````

````python
# Crear un rango de valores para x
x = np.linspace(-10, 10, 400)
````

````python
# Graficar las funciones
plt.plot(x, f(x), 'r--', label=r'$f(x) = x^2 + 2x + 3$')  # Lnea roja segmentada
plt.plot(x, g(x), 'g-', label=r'$g(x) = -x^2 + 1$')       # Lnea verde continua

# Aadir leyenda, ttulo, etiquetas y cuadrcula
plt.legend()
plt.title('Grfica de las funciones f(x) y g(x)')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.grid(True)

# Mostrar la grfica
plt.show()
````

## Ejercicio 4

Grafique una estrella de 5 puntas, con borde de color naranjo, y rellena de color amarillo.

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Funcin para calcular los vrtices de una estrella de 5 puntas
def estrella_5_puntas(radio_exterior, radio_interior):
    angulos = np.linspace(0, 2 * np.pi, 10, endpoint=False)
    vertices = []

    for i, angulo in enumerate(angulos):
        if i % 2 == 0:  # Vrtices de la parte exterior
            vertices.append((radio_exterior * np.cos(angulo), radio_exterior * np.sin(angulo)))
        else:  # Vrtices de la parte interior
            vertices.append((radio_interior * np.cos(angulo), radio_interior * np.sin(angulo)))

    vertices.append(vertices[0])  # Cerrar la estrella
    return np.array(vertices)
````

````python
# Parmetros de la estrella
radio_exterior = 1
radio_interior = 0.5
````

````python
# Obtener los vrtices
vertices = estrella_5_puntas(radio_exterior, radio_interior)
````

````python
# Crear la figura y el eje
plt.figure(figsize=(6, 6))
plt.fill(vertices[:, 0], vertices[:, 1], 'yellow', edgecolor='orange', linewidth=3)

# Aadir ttulo, etiquetas y cuadrcula
plt.title('Estrella de 5 puntas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)

# Ajustar los lmites de la grfica
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)

# Quitar ejes
plt.gca().set_aspect('equal', adjustable='box')
plt.xticks(np.linspace(-1, 1, 5))  # Etiquetas en el eje X
plt.yticks(np.linspace(-1, 1, 5))  # Etiquetas en el eje Y

# Mostrar la estrella
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Crear el DataFrame replicando la tabla de la imagen
data = {
    'nombre': ['Jos Luis Martnez Izquierdo', 'Rosa Diaz Diaz', 'Javier Garca Snchez',
               'Carmen Lpez Pinzn', 'Marisa Lpez Collado'],
    'edad': [18, 32, 24, 35, 46],
    'sexo': ['H', 'M', 'H', 'M', 'M'],
    'peso': [85.0, 65.0, None, 65.0, 51.0],
    'altura': [1.79, 1.73, 1.81, 1.70, 1.58],
    'colesterol': [182.0, 232.0, 191.0, 200.0, 148.0]
}

df = pd.DataFrame(data)
````

````python
# (b) Mostrar slo a las mujeres y calcular el promedio de su colesterol
mujeres = df[df['sexo'] == 'M']
print(mujeres)
promedio_colesterol_mujeres = mujeres['colesterol'].mean()
print(f'Promedio de colesterol de mujeres: {promedio_colesterol_mujeres}')
````
